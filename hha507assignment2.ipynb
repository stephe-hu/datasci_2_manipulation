```python
import pandas as pd
import numpy as np
import re
```

# Loading file

```python
df = pd.read_csv('https://raw.githubusercontent.com/stephe-hu/datasci_2_manipulation/main/data/healthcare_data_cleaning.csv')
df
```

# 1. Data Cleaning with Pandas

## Identify and handle missing values in the dataset

```python
# Identifying and replace 'missing' values in columns NaN
df = df.replace('missing', np.nan)
# Drop rows with any NaN values
df.dropna(inplace=True)
df
```

## Remove any duplicate rows and columns, if they exist

```python
# Identify and remove duplicate rows
df.drop_duplicates(inplace=True)
df
```

## Clean column names

```python
# Remove leading and trailing white space, convert to lowercases, and replace spaces with underscores for columns
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df
```

# 2. Data Transformation

## Create new columns based on existing ones

```python
# Convert 'patient_age' to int and then create new column 'age_group'
df['patient_age'] = df['patient_age'].astype(int)
age_bins = [0, 30, 50, np.inf]
age_labels = ['Young', 'Middle-aged', 'Senior']
df['age_group'] = pd.cut(df['patient_age'], bins=age_bins, labels=age_labels, right=False)
df
```

```python
# Convert 'average_bp' to int and then create new column 'blood_pressure_status'
df['average_bp'] = df['average_bp'].astype(int)
threshold_high = 140
df['blood_pressure_status'] = df['average_bp'].apply(lambda x: 'High' if x >= threshold_high else 'Normal')
df
```

## Aggregate data using groupby and compute summary statistics.

```python
# Data aggregation to count patient count in each state
state_pop = df.groupby('state_of_residence').size().reset_index(name='patient_count')
state_pop
```

```python
# Compute summary statistics
health_stats = df[['average_heart_rate','average_bp','height_(in_cm)','weight_(in_kg)']]
health_stats.describe()
```

```python
# Use pivot tables or cross-tabulations for multi-dimensional analysis
pivot_table = pd.pivot_table(df, index = 'gender', columns = 'disease_diagnosed', values = ['average_bp', 'average_heart_rate'], aggfunc = 'mean')
pivot_table
```
