```python
import pandas as pd
import numpy as np
import re
```

# Loading file

```python
df = pd.read_csv('https://raw.githubusercontent.com/hantswilliams/HHA_507_2023/main/WK2/data/healthcare_data_cleaning.csv')
df
```

# Data Cleaning with Pandas

```python
# Identifying and replace 'missing' values in columns NaN
df = df.replace('missing', np.nan)
# Drop rows with any NaN values
df.dropna(inplace=True)
df
```

```python
# Identify and remove duplicate rows
df.drop_duplicates(inplace=True)
df
```

```python
# Remove leading and trailing white space, convert to lowercases, and replace spaces with underscores for columns
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df
```
